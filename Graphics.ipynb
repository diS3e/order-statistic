```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
sns.set_theme()
```

```python
data1 = pd.read_csv("data.csv", sep=";", header=None)
data2 = pd.read_csv("data2.csv", sep=";", header=None)
```

```python
fig, ax = plt.subplots() 
fig.set_size_inches(13, 8)
g = sns.lineplot(data1[0], 1 - data1[1])
plt.yscale('log')

ax.set_title("Вероятность ошибки для кода (16, 11)", fontsize=15) 
plt.xlabel("Сигнал/шум на бит, дБ", fontsize=13) 
plt.ylabel("Вероятность", fontsize=13)

plt.savefig('1.png')
```

```python
fig, ax = plt.subplots() 
fig.set_size_inches(13, 8)
sns.lineplot(data2[0], 1 - data2[1])
plt.yscale('log')
ax.set_title("Вероятность ошибки для кода (32, 21)", fontsize=15) 
plt.xlabel("Сигнал/шум на бит, дБ", fontsize=13) 
plt.ylabel("Вероятность", fontsize=13)

plt.savefig('2.png')
```

```python
fig, ax = plt.subplots() 
fig.set_size_inches(13, 8)
sns.lineplot(data2[0], 1 - data2[1], label="(32, 21)")
sns.lineplot(data1[0], 1 - data1[1], label="(16, 11)")
plt.yscale('log')
ax.set_title("Вероятность ошибки", fontsize=15) 
plt.xlabel("Сигнал/шум на бит, дБ", fontsize=13) 
plt.ylabel("Вероятность", fontsize=13)


plt.savefig('3.png')
```
